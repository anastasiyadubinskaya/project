```python
# Ex 1.2: Define the dimensions of the crop, make sure that min is smaller than max and also that min is larger than 0 
# and max is smaller than the size of the image for every dimension. If not, print a message explaining the error.

# Example crop dimensions, modify them to test your code

crop_min_h = 100
crop_max_h = 300
crop_min_w = 800
crop_max_w = 1000

hasError = False

if crop_min_h > crop_max_h:
    print("Minimum height is larger than maximum height.")
    hasError = True

if crop_min_w > crop_max_w:
    print("Minimum width is larger than maximum width.")
    hasError = True

if crop_min_h < 0:
    print("Minimum height should be larger than 0.")
    hasError = True

if crop_min_w < 0:
    print("Minimum width should be larger than 0.")
    hasError = True

if crop_max_h > img.size[1]:
    print("Maximum height should be smaller than the height of the image.")
    hasError = True

if crop_max_w > img.size[0]:
   print("Maximum width should be smaller than the width of the image.")
   hasError = True


```

```python
# Ex 1.3: Generate the crop array into a new variable, use NumPy array slicing

if not hasError:
  crop_arr = img_arr[crop_min_h:crop_max_h, crop_min_w:crop_max_w, :] 
 
```

```python
# Display the cropped image

plt.imshow(crop_arr)
```

```python
# Ex 1.4: Saving the cropped image into a new file, set the filename of the new image

crop_img = Image.fromarray(crop_arr)

file_name = "cropped_image" #Set a file name, use a string

crop_img.save(f"../data/{file_name}.png")
```
